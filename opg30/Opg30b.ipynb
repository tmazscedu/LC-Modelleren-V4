```python
# Inladen bibliotheken en hulpfiles
import matplotlib.pyplot as plt
%matplotlib inline   
## Do not change unless you know what you need
```

```python
# constanten en variabelen (vul zelf aan)
t = 0
dt = 1
m = 3.04E6
mvl = 12.9E3
Fstuwvol = 34.8E6
Fstuw = 0
g=-9.81
Fres = 0
Fz = 0



# lijstvariabelen (alleen als je later grafieken wilt maken)
tlijst = []
mlijst = []
Fstuwlijst = []
Freslijst = []
Fzlijst = []



```

```python
# loop opzetten tot wanneer en berekeningen toevoegen
while t < 160:
    Fz = m * g
    if t>134:
        Fstuw = Fstuwvol * 0.8
    else:
        Fstuw = Fstuwvol
    Fres = Fstuw + Fz
    m = m - mvl
    t = t + dt

    ######  sla de berekende waarden op in de lijsten
    tlijst.append(t)
    mlijst.append(m)
    Freslijst.append(Fres)
    Fstuwlijst.append(Fstuw)
    Fzlijst.append(Fz)
    




```

```python
# laat de grafieken zien 
plt.plot(tlijst,Freslijst)
plt.plot(tlijst,Fstuwlijst)
plt.show()
plt.plot(tlijst,mlijst)
plt.show()
plt.plot(tlijst,Fzlijst)
plt.show()
```
