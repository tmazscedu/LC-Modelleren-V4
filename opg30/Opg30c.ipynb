```python
# Inladen bibliotheken en hulpfiles
import matplotlib.pyplot as plt
%matplotlib inline   
## Do not change unless you know what you need
```

```python
# constanten en variabelen (vul zelf aan)
t = 0
dt = 0.1
m = 2.945973E6
mvl = 12.9E3*0.1
Fstuwvol = 34.8E6
Fstuw = 0
g=-9.81
Fres = 0
Fz = 0
k = 0  #1.97
v = 0


# lijstvariabelen (alleen als je later grafieken wilt maken)
tlijst = []
mlijst = []
Fstuwlijst = []
Freslijst = []
Fzlijst = []
vlijst = []
alijst = []



```

```python
# loop opzetten tot wanneer en berekeningen toevoegen
while t < 135:
    Fz = m * g
    Fstuwvol = Fstuwvol * 1.00013 
    if t>135:
        Fstuw = Fstuwvol * 0.8
    else:
        Fstuw = Fstuwvol
    a = Fres/m
    dv = a*dt
#    if t<7:
#        dv = 0
    v=v+dv
    Fw = k * v * v * -1
    Fres = Fstuw + Fz + Fw
    m = m - mvl
    t = t + dt

    ######  sla de berekende waarden op in de lijsten
    tlijst.append(t)
    mlijst.append(m)
    Freslijst.append(Fres)
    Fstuwlijst.append(Fstuw)
    Fzlijst.append(Fz)
    vlijst.append(v)
    alijst.append(a)
    print(t-0.1,a,v,Fstuw)

    


```

```python
# laat de grafieken zien 
plt.plot(tlijst,Freslijst)
plt.plot(tlijst,Fstuwlijst)
plt.show()
plt.plot(tlijst,mlijst)
plt.show()
plt.plot(tlijst,Fzlijst)
plt.show()
plt.plot(tlijst,vlijst)
plt.show()
plt.plot(tlijst,alijst)
plt.show()
print(t,a,v)
print(Fstuwvol)
print(m)
```
